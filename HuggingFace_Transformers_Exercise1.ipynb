## 1. Load Model & Tokenizer

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "distilgpt2"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

tokenizer.pad_token = tokenizer.eos_token
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

print(f"Model : {model_name}")
print(f"Params: {sum(p.numel() for p in model.parameters()):,}")
print(f"Device: {device}")
```

## 2. Tokenization
`text → tokens → token IDs`

```python
text = "The future of AI is"
inputs = tokenizer(text, return_tensors="pt").to(device)

print("Text      :", text)
print("Tokens    :", tokenizer.tokenize(text))
print("Token IDs :", inputs["input_ids"].tolist()[0])
```

## 3. Inference & Generation Parameters


```python
def generate(temp, top_p, max_new_tokens):
    output = model.generate(
        inputs["input_ids"],
        max_new_tokens=max_new_tokens,
        temperature=temp,
        top_p=top_p,
        do_sample=True,
        pad_token_id=tokenizer.eos_token_id
    )
    return tokenizer.decode(output[0], skip_special_tokens=True)


configs = [
    {"temperature": 0.6, "top_p": 0.8, "max_new_tokens": 30},
    {"temperature": 0.6, "top_p": 0.9, "max_new_tokens": 50},
    {"temperature": 0.8, "top_p": 0.8, "max_new_tokens": 30},
    {"temperature": 0.8, "top_p": 0.9, "max_new_tokens": 50},
    {"temperature": 1.0, "top_p": 0.8, "max_new_tokens": 30},
    {"temperature": 1.0, "top_p": 0.9, "max_new_tokens": 50},
    {"temperature": 1.2, "top_p": 0.8, "max_new_tokens": 50},
    {"temperature": 1.2, "top_p": 0.9, "max_new_tokens": 100},
]

for i, cfg in enumerate(configs, 1):
    print(f"\nOutput {i} | {cfg}")
    print(generate(cfg["temperature"], cfg["top_p"], cfg["max_new_tokens"]))
```
